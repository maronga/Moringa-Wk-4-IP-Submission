## 4. Data Cleaning

```
#We'll first drop the cars column because it's redundant with the bluecar column
#This is done while also dropping the other columns which are irrelevant in this
#data analysis
df.drop(['Cars','ID','Geo point','Displayed comment','year','month'], axis = 1, inplace = True) 
df.head()
```

```
#Rename the column name public name to station name for easier readability
#
df.rename(columns={"Public name": "Station"}, inplace=True)
df.head()
```

## 5. Filtering and Sorting

### Blue Car Analysis 


```
#Inorder to answer the research questions, we will need to filter the data
#The first step is to create a new column where we'll find the difference in the 
#number of blue cars at the station/counter
df['Bluecar_Difference'] = df['Bluecar counter'].diff()
df.head()
#If the row value in the bluecar_difference column is +ve, it means the car was 
#picked and if it's -ve, it means the car was returne
```

```
# Identify the most popular hour of the day for picking up a shared electric car 
#(Bluecar) in the city of Paris over the month of April 2018
paris = df[df['City'] =='Paris'] 
paris.head()
paris[paris['Bluecar_Difference'] > 0].groupby('hour')['hour'].count().sort_values(ascending= False)
#The most popular hour is 3:00 hours
```

```
# Most popular hour for returning vehicles
paris[paris['Bluecar_Difference'] < 0].groupby('hour')['hour'].count().sort_values(ascending= False)
#the most popular hour is 21:00 hours
```

```
# What station is the most popular overall
#
paris[(paris['Kind'] == 'STATION') & (paris['Status'] == 'ok')].groupby('Station')['Station'].count().sort_values(ascending= False)
#paris/porte de Montrouge/8 is the most popular overall
```

```
# What station is the most popular at the most popular picking hour?
paris[(paris['Kind'] == 'STATION') & (paris['Status'] == 'ok') & (paris['hour'] == 3)].groupby('Station')['Station'].count().sort_values(ascending= False)

```

```
# What postal code is the most popular for picking up Blue cars?
paris[paris['Bluecar_Difference'] > 0].groupby('Postal code')['Postal code'].count().sort_values(ascending= False)
#75015 is the most popular postal code which belongs to	Paris/Suffren/2 station,
# therefore it's not the same as the most popular station Paris/Porte de Montrouge/8 
```

```
# What postal code is the most popular for picking blue cars at the most popular 
# picking hour
paris[(paris['Kind'] == 'STATION') & (paris['Status'] == 'ok') & (paris['hour'] == 3)].groupby('Postal code')['Postal code'].count().sort_values(ascending= False)
#75016 is the most popular postal code
```
